# 1. Multi-class and Multi-Label Classification Using Support Vector Machines

### 1(a). Download the Anuran Calls (MFCCs) Data Set from: https://archive.ics.uci.edu/ml/datasets/Anuran+Calls+%28MFCCs%29.Choose 70% of the datarandomly as the training set.

```python
csv_data = pd.read_csv('Frogs_MFCCs.csv')
csv_data
```

```python
training_data = csv_data.sample(frac=0.70, random_state=99)
test_data = csv_data.loc[~csv_data.index.isin(training_data.index), :]
training_data.reset_index(drop=True,inplace=True)
test_data.reset_index(drop=True,inplace=True)
test_data
```

### 1(b). Each instance has three labels: Families, Genus, and Species. Each of the labels has multiple classes. We wish to solve a multi-class and multi-label problem. One of the most important approaches to multi-class classification is to train a classifier for each label. We first try this approach:

### 1(b)(i) Research exact match and hamming score/ loss methods for evaluating multilabel classification and use them in evaluating the classifiers in this problem.

```python
from sklearn.metrics import hamming_loss

from sklearn import svm

y_training_1 = training_data[["Family"]]
y_training_2 = training_data[["Genus"]]
y_training_3 = training_data[["Species"]]
x_training = training_data.drop(["Family", "Genus", "Species", "RecordID"], axis=1)

y_test_1 = test_data[["Family"]]
y_test_2 = test_data[["Genus"]]
y_test_3 = test_data[["Species"]]
x_test = test_data.drop(["Family", "Genus", "Species", "RecordID"], axis=1)


clf = svm.SVC(gamma='auto')
model = clf.fit(x_training, y_training_1)

y_pred_1 = model.predict(x_test)

hamming_loss(y_test_1, y_pred_1)
```

```python
pred_1 = pd.DataFrame(data=y_pred_1, columns=y_training_1.columns)
sum = 0
for i in range(0,len(pred_1)):
    if pred_1.values.tolist()[i][0] == y_training_1.values.tolist()[i][0]:
        sum +=1
exact_match = sum/len(pred_1)
exact_match
```

```python
clf = svm.SVC(gamma='auto')
model = clf.fit(x_training, y_training_2)

y_pred_2 = model.predict(x_test)

hamming_loss(y_test_2, y_pred_2)
```

```python
pred_2 = pd.DataFrame(data=y_pred_2, columns=y_training_2.columns)
sum = 0
for i in range(0,len(pred_2)):
    if pred_2.values.tolist()[i][0] == y_training_2.values.tolist()[i][0]:
        sum +=1
exact_match = sum/len(pred_2)
exact_match
```
